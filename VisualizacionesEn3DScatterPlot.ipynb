```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from google.colab import files
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/data/pinguinos.csv')
df = df.dropna()
df
```

```python
#crear la figura en matplotlib
fig= plt.figure(figsize=(10,7))
ax=plt.axes(projection='3d')
ax.scatter3D(
    df['culmen_length_mm'],
    df['culmen_depth_mm'],
    df['flipper_length_mm'],
    color='green'
)
ax.set_xlabel('Longitud del Pico')
ax.set_ylabel('Profundidad del Pico')
ax.set_zlabel('Longitud de las Aletas')
plt.show()
# en este gráfico la intensidad del color verde  es para diferenciar la profundidadd
```

```python
# Ahora vamos a hacer el mismo gráfico con los mismos datos pero en Plotly
fig = px.scatter_3d(
    df,
    x='culmen_length_mm',
    y='culmen_depth_mm',
    z='flipper_length_mm',
    color='species',
    symbol='species'
)
fig.show()
```
